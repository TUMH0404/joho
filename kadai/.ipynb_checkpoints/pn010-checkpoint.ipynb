- 血圧の分類を表す関数を作り，データ収集をする。

分類|収縮期||拡張期
---|----|----
至適血圧|<120|かつ|<80
正常血圧|<130|かつ|<85
正常高血圧|130〜139|または|85〜89
Ⅰ高血圧|140〜159|または|90〜99
Ⅱ度高血圧|160〜169|または|100〜109
Ⅲ度高血圧|≧180|または|≧110
収縮期高血圧|≧140|かつ|<90

```python

def BP(sbp,dbp):
    '''
    最高血圧と最低血圧から血圧値を分類する。
    sbp:最高血圧，dbp:最低血圧
    '''
    ssn = {5:"Ⅲ度高血圧",4:"Ⅱ度高血圧",3:"Ⅰ度高血圧",2:"正常高血圧",1:"正常血圧",0:"至適血圧",7:"収縮期高血圧",-1:"低血圧"}

    if not sbp<100:
    
        ss = sum([sbp>=180,sbp>=160,sbp>=140,sbp>=130,sbp>=120])
        dd = sum([dbp>=110,dbp>=100,dbp>=90,dbp>=85,dbp>=80])
    
        if ss==dd and ss<=1:
            result = ssn[ss]

        elif ss>=3 and dd<3:
            result = ssn[7]
            
        else:
            num = [ss,dd] 
            num.sort()  ## ソートして大きい方を選択する
            result = ssn[num[1]]

    else:
        result = ssn[-1]
       
    return f"成功_{result}"




import datetime
import pickle
import os

def main():
    """
    血圧の結果をIDと紐づけて保存していく
    """
    
    if 'BP.pickle' in glob.glob("*"):
        with open("BP.pickle","rb") as d:
            BPresult = pickle.load(d)
    else:
        BPresult ={}
    
    
    while(True):
        ID = input("IDは？:\n")
        Sbp = int(input("上の血圧：\n"))
        Dbp = int(input("\n下の血圧：\n"))
        time = datetime.datetime.now()
        BPresult.update({ID:(Sbp,Dbp,time.year,time.month,time.day,time.hour,time.minute)})
        if Sbp < Dbp:
            print("入力間違いしていませんか？\n入力し直してください")
        
        else:
            #print(f"あなたの血圧{Sbp}/{Dbp}mmHg")
            print(BP(Sbp,Dbp))
    
        key = input("入力終了? Y[yes]/n[no] \n")
        if key == "Y":
            
            with open('BP.pickle',"wb") as d:
                pickle.dump(BPresult,d)
            
            break
        
        else:
            pass
    
    print(f"今回入力したデータ\n")
    for i in list(BPresult.keys()):
        d0 = i
        d1 = BPresult[i][0]
        d2 = BPresult[i][1]
        y = BPresult[i][2]
        m = BPresult[i][3]
        d = BPresult[i][4]
        hh = BPresult[i][5]
        mm = BPresult[i][6]
        
        print(f"{y}-{m}-{d} {d}:{hh}:{mm}  {d0}  {d1}/{d2} mmHg___{BP(d1,d2)}")
    

    ####
```
