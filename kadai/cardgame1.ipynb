## 下準備

````python
## 必要なライブラリをインポートする
#########################
import random
# 以上 #

#########################
def cardset(joka=0):
    '''
    トランプカードを作って混ぜる。
    jokaでジョーカーの枚数を指定する
    '''
# 格納する空リスト
    card = []
## トランプカードの種類
# マーク
    mark =["spade","heart","diamond","club"]
# 数
    num = ["A","K","Q","J","X","9","8","7","6","5","4","3","2"]
# ジョーカー
    Joker = ["j1","j2"]

# それぞれ組み合わせる
    for i in mark:
        for k in num:
            card.append(f"{k}_{i}")

# ジョーカーを追加する
    if joka==1 or joka==2:
        card.append(Joker[joka-1])
    else:
        pass

# 混ぜる
    for i in range(5):
        random.shuffle(card)
        
    return card
# 以上 #


#########################
def dealcard(Gcard,N):
    '''
    トランプを配る。
    Gcard:トランプ
    N:配られる人数もしくは，リストに納められた文字型の名前
    '''
    cc = {}
    if isinstance(N,int):
        NN = N
        for u in range(N):
            cc.update({f"p{u}":[]})
    
    elif isinstance(N,list):
        NN = len(N)
        random.shuffle(N)
        for u in N:
            cc.update({str(u):[]})
    
    else:
        NN = 4
        for u in range(4):
            cc.update({f"p{u}":[]})
    
    
    p = list(cc.keys())

    maisuu = int(len(Gcard)//NN)# 配る枚数
    nokori = len(Gcard)%NN      # 残り

    for kubaru in range(maisuu):
        #random.shuffle(p)
        a = random.sample(Gcard,NN)
        for i,k in zip(p,a):
            cc[i].append(k)
            Gcard.remove(k) 

    ## 残りのカード誰に配る？
    morauhito = random.sample(p,nokori)
    for u,y in zip(morauhito,Gcard):
        cc[u].append(y)
    
    return cc
# 以上 #

#################################
def baba3(A):
    '''
    配られたカードを整理する
    '''
    c2 = A[:] ## 深いコピー
    ## 並べ替え
    c2.sort()
    ## 数字を取り出す
    c0 = [i[0] for i in c2]
    # 集合してリスト化する
    c1 = list(set(c0))
    # ソートする
    c1.sort()
    
    # 一枚もののカードを格納する空リストを作る。
    single =[]
    
    # 数字を回して検索する
    for i in c1:
        # 念の為ソートする
        c2.sort()
        
        # 同じものが複数ある時の処理。
        # 一枚になれば，singleに追加する。
        gg0 = c0.count(i)//2
        gg1 = c0.count(i)%2
        if gg0==1 and gg1 == 0:  # 偶数だけ捨てる
            del c2[:2]
            
        elif gg0==2 and gg1==0:   # 偶数だけ捨てる
            del c2[:4]
            
        elif gg0==1 and gg1==1:
            del c2[:2]
            single.append(c2[0]) #組み合わせがないカードだけを格納
            del c2[0]
            
        elif gg0==0 and gg1 ==1:
            single.append(c2[0])  #組み合わせがないカードだけを格納
            del c2[0]
            
    return single
# 以上 #


###################
def pickup(me,you):
    '''
    カードの受け渡し
    me:取る人
    you:取られる人
    '''
    
    random.shuffle(you)             # youはカードを混ぜる
    if len(you)==0:
        print("youの長さがゼロです")
        
        return 0,0
    
    
    else:
        a = list(range(len(you))) 
        ii = random.sample(a,1)[0] # 好きなカードを選ぶ 
    
        me.append(you[ii])         # とったカードと手元のカードを合わせる
        del you[ii]                # カードを渡す
        
        return baba3(me),you       # 結果
# 以上 # 
````

## 実行セル

````python
## ババ抜きしてみよう ##
## 実行するためのセル  ##
cc =dealcard(cardset(joka=1),["tomo","kobay","takami","shinohara"])
## カードを整理する(揃う数字を取り除いく) ##
for i in cc:
    cc[i] = baba3(cc[i])

## その中身を確認する ##   
print(cc)


## いよいよゲーム開始! ##
## 一瞬で終わるので見逃さないように ##
## 誰が勝つでしょう？？  ##
goal = []
times = 0
pp=cc.copy()
while 1:
    menbers = list(pp.keys())
    for ite,i in enumerate(menbers):
        if ite+1 == len(menbers):
            me = menbers[ite]
            you = menbers[0]

        else:
            me = menbers[ite]
            you =menbers[ite+1]
        
        if len(pp[me])==0:
            continue
    
        else:
            A,B = pickup(pp[me],pp[you])
            if A!=0 and B!=0:
                pp[me] = A
                pp[you] = B
            
            else:
                continue
        print(pp)
    

    for u in list(pp.keys()):
        if len(pp[u])==0:
            goal.append(u)
            del pp[u]
        
        else:
            continue
    
    #print(f"test:{len(pp)}")
    times+=1
    
    if len(pp)==1:
        goal.extend(list(pp.keys()))
        print(goal)
        break

## 結果の表示 ##
print("\n"*2)
print(f"{times}回")
print("\n"*2)
print("順位")
for y,i in enumerate(goal):
    print(f"第{y+1}位:{i}")
````
